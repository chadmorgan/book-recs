```python
import pandas as pd
from matplotlib import pyplot as plt
```

### Book metadata

```python
books = pd.read_csv('goodbooks-10k-master/books.csv')
books.head()
```

```python
books['first_author'] = books.authors.apply(lambda x: x.split(',')[0])
```

```python
len(books.drop_duplicates('first_author'))
```

```python
plt.hist(books['average_rating'],50)
plt.show()
```

```python
books.sort_values('average_rating',ascending=False).head(30)[['goodreads_book_id','title','first_author','average_rating']]
```

```python
books.sort_values('average_rating').head()[['goodreads_book_id','title','first_author','average_rating']]
```

```python
books[books['goodreads_book_id']==9814682][['ratings_1','ratings_2','ratings_3','ratings_4','ratings_5']
                                          ].values[0].tolist()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.plot(books[books['goodreads_book_id']==9814682][
    ['ratings_1','ratings_2','ratings_3','ratings_4','ratings_5']].values[0].tolist()
        ,'-o')
plt.title('Game of thrones')
plt.subplot(122)
plt.plot(books[books['goodreads_book_id']==34][
    ['ratings_1','ratings_2','ratings_3','ratings_4','ratings_5']].values[0].tolist()
        ,'-o')
plt.title('Fellowship of the Ring')
plt.legend()
plt.show()
```

### Ratings

```python
ratings = pd.read_csv('goodbooks-10k-master/ratings.csv')
ratings.head()
```

```python
len(ratings.drop_duplicates('user_id'))
```

```python
user_avg_rating = ratings.groupby('user_id')['rating'].mean()
```

```python
plt.hist(user_avg_rating,50)
plt.show()
```

### Tags

```python
book_tags = pd.read_csv('goodbooks-10k-master/book_tags.csv')
book_tags.head()
```

```python
tags = pd.read_csv('goodbooks-10k-master/tags.csv')
tags.tail()
```

```python
# the hobbit
book_tags[book_tags['goodreads_book_id']==5907].merge(tags,on='tag_id').sort_values('count',ascending=False).head(15)
```

```python
# fellowship of the ring
book_tags[book_tags['goodreads_book_id']==34].merge(tags,on='tag_id').sort_values('count',ascending=False).head(10)
```
